Load experimental data

```python
import matplotlib.pyplot as plt
import numpy as np
A0 = np.genfromtxt(fname='data_0.txt')
A1 = np.genfromtxt(fname='data_1.txt')
A2 = np.genfromtxt(fname='data_2.txt')
A3 = np.genfromtxt(fname='data_3.txt')
A6 = np.genfromtxt(fname='data_6.txt')
A12 = np.genfromtxt(fname='data_12.txt')
A24 = np.genfromtxt(fname='data_24.txt')
A26 = np.genfromtxt(fname='data_26.txt')
A28 = np.genfromtxt(fname='data_28.txt')
A30 = np.genfromtxt(fname='data_30.txt')
# print(A0), print(A1), print(A2), print(A3), print(A6), print(A12), print(A24), print(A26), print(A28), print(A30)
```

Define quantile percentages

```python
# q=[0,0.2,0.4,0.6,0.8,1]
q=[0,0.1,0.25,0.5,0.75,0.9,1]
```

Calculate diameters corresponding to quantiles

```python
q0 = np.quantile(A0, q)
q1 = np.quantile(A1, q)
q2 = np.quantile(A2, q)
q3 = np.quantile(A3, q)
q6 = np.quantile(A6, q)
q12 = np.quantile(A12, q)
q24 = np.quantile(A24, q)
q26 = np.quantile(A26, q)
q28 = np.quantile(A28, q)
q30 = np.quantile(A30, q)
```

Generate table of all quantile data at all times

```python
Q = np.array([q0, q1, q2, q3, q6, q12, q24, q26, q28, q30],float)
print(Q)
```

Visualization

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 12})

Q0 = Q[:,0]; Q1 = Q[:,1]; Q2 = Q[:,2]; Q3 = Q[:,3]; Q4 = Q[:,4]; Q5 = Q[:,5]; Q6 = Q[:,6];
T = [0, 1, 2, 3, 6, 12, 24, 26, 28, 30]
Tlab = ['0', '1', '2', '3', '6', '12', '24', '26', '28', '30']

plt.plot(T, Q0,'o', label="0", color="black", ls="-", lw="0.2")
plt.plot(T, Q1,'o', label="0.1", color="black", ls="-", lw="0.2")
plt.plot(T, Q2,'o', label="0.25", color="black", ls="-", lw="0.2")
plt.plot(T, Q3,'o', label="0.5", color="black", ls="-", lw="2")
plt.plot(T, Q4,'o', label="0.75", color="black", ls="-", lw="0.2")
plt.plot(T, Q5,'o', label="0.9", color="black", ls="-", lw="0.2")
plt.plot(T, Q6,'o', label="1", color="black", ls="-", lw="0.2")

plt.xticks(T, Tlab)
# plt.legend(loc='lower center')
plt.xlabel('Time [h]')
plt.ylabel('Diameter [m]')
plt.savefig('CountingQuantiles.png')
plt.show()
```

## For initial time generate initial data for equidistant intervals

```python
D0 = np.genfromtxt(fname='data_0.txt')
V0=D0**3*np.pi/6
K=[2,4,8,9,11]
K=range(12)
K=[12, 18]
K=range(42)
for k in K:
    L=np.linspace(0,300,k);
    
    # numpy.histogram(a, bins=10, range=None, normed=None, weights=None, density=None)    
    #    hist, bin_edges = np.histogram(D0, L)
    hist, bin_edges = np.histogram(D0, L, weights=V0)    
    # print(k, ':', hist); print()
    
    Vsel=[]
    for k in range(len(L)-1):
        MASK = (D0 > L[k]) & (D0 <= L[k+1])
        E=V0*MASK
        Vsel.append(sum(E))

    Lavg=(L[0:-1]+L[1:])/2
    dL = np.diff(L)-2
    plt.plot(Lavg, Vsel, color='#000000', marker='o', markersize=8, lw=0.5)
    plt.bar(Lavg, Vsel, dL, color='#777777')
    plt.xlabel('Diameter [m]')
    plt.ylabel('Volume [m^3]')
    plt.title('k='+str(k))
    
    plt.ylim([0, 1e7])
    plt.grid(axis='y', alpha=1)
    plt.savefig('C0_V'+ str(k) +'.png')    
    
    plt.show()    
    print('k : ', k, '; L : ', L, '; Vol : ', Vsel)
    print()
```

Generate set of histograms for each time

```python
dict_B = {}
for t in Tlab:
    dict_B["B{0}".format(t)] = np.genfromtxt(fname='data_'+ t +'.txt')

# L=np.linspace(0,300,11);
L=np.linspace(0,300,21);

import matplotlib
matplotlib.rcParams.update({'font.size': 12})

for t in Tlab:
    X = dict_B.get("B"+t)
    plt.hist(x=X, bins=L, color='#777777', rwidth=0.8)
   
    plt.xlabel('Diameter [m]')
    plt.ylabel('Count')
    plt.title('T=' + t)
    # plt.ylim([0, 40])
    plt.ylim([0, 30])    
    plt.grid(axis='y', alpha=1)
    plt.savefig('ch_21_t'+ t +'.png')

    plt.show()
```

## Cumulative distributions

```python

LINESTYLE = ['solid', 'dashed', 'dashdot', 'dotted'] 
MARKER = ['^', 's', 'x', '.']
Tsub = ['0', '3', '12', '24']

k=-1
for t in Tsub:    
    k=k+1
    D = dict_B.get("B"+t)
    D=np.sort(D) # Sort     
    V=D**3*np.pi/6 # Calculate volumes
    W = np.cumsum(V)/sum(V) # cumulative distribution function
    # 
    plot1 = plt.figure(1)

    spl = UnivariateSpline(D, W)
    spl.set_smoothing_factor(0.005)
    spl.set_smoothing_factor(0.01)
    
    xx = np.linspace(min(D),max(D),100000)
    # xx = np.linspace(0,300,100000)
    
    yy = spl(xx)
    plt.plot(xx, yy, 'b', lw = 2, linestyle = LINESTYLE[k])    
    plt.plot(D, W, color='#000000', marker = MARKER[k], linestyle = 'None')    
    
    ###################################################
    ### non-smooth differentiation by hand
    # DW=np.diff(W)
    # DD=np.diff(D)
    # DW_divided_by_DD=np.divide(DW, DD)
    # plot2 = plt.figure(2)
    # plt.plot(D[0:-1], DW_divided_by_DD,color='#000000', linestyle = LINESTYLE[k])  
    
    plot2 = plt.figure(2)
    Wder = spl.derivative()
    xx = np.linspace(min(D),max(D),100000)
    
    yy_der = Wder(xx)
    plt.plot(xx, yy_der, 'b', lw = 2, linestyle = LINESTYLE[k])
    plt.plot(D, Wder(D), color='#000000', marker = MARKER[k], linestyle = 'None')
    
    
plot1 = plt.figure(1)
plt.xlabel('Diameter [m]')
plt.ylabel('Cumulative distribution')
plt.grid(axis='y', alpha=1)
plt.savefig('distribution-cumulative.png')


plot2 = plt.figure(2)
plt.xlabel('Diameter [m]')
plt.ylabel('Density distribution')
plt.grid(axis='y', alpha=1)
plt.savefig('distribution-density.png')
plt.ylim([0, 0.025])
```

## Volume sums

```python
Vtab = []
k=-1
for t in Tlab:    
    k=k+1
    D=dict_B.get("B"+t)
    D=np.sort(D) # Sort
    D=D/1000
    V=D**3*np.pi/6 # Calculate volumes
    vsum=sum(V)
    Vtab.append(vsum)
    print(t+': \t'+str(vsum))
    
print(Tlab, Vtab)

# plt.plot(Tlab, Vtab, '.', color="black", ls="-", lw="1")
plt.plot(T, Vtab, '.', color="#000000", ls="-", lw="0.1")
plt.xticks(T, Tlab)
plt.xlabel('Time [h]')
plt.ylabel('Volume [mm^3]')
plt.savefig('volumesum.png')
```
